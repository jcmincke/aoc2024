```python
import os 
import pandas as pd
```

```python
df = pd.read_csv('input.txt', sep='   ', header=None, names=['col1', 'col2'])
df.head()
```

```python
df['col1'] = sorted(df['col1'])
df['col2'] = sorted(df['col2'])
df.head()
```

```python
# part 1
dist = sum(abs(df.col1-df.col2))
dist
```

```python
# part 2
# let's compare each row in col1 to each row of col2
df.col1.apply(lambda x: df.col2 == x)
```

```python
# answer
(df.col1.apply(lambda x: df.col2 == x).sum(1)*df.col1).sum()
```

```python
# explanation
df['col3'] = df.col1.apply(lambda x: df.col2 == x).sum(1) # sum on axis 1 = nb of times the value on that row is present in col2
df.head(10)
```

```python
# let's have a look at which ones are present in both cols
df[df.col3 != 0]
```

```python
(df.col1*df.col3).sum()
```
